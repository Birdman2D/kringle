```python
import requests as req
import json
from urllib.request import urlopen

status = False
rid = 'dcf77635-6971-4503-a8ea-87bcabf0c07c'
f = open( "file.txt", "r" ).read()
lines = f.split("\n")
lineno = ""
while not status:
    for line in lines:
        lineno = line
        u = f"https://doorpasscode.kringlecastle.com/checkpass.php?i={line}&resourceId={rid}"
        url = urlopen(u).read()
        obj = json.loads(url.decode('utf-8'))
        success = obj["success"]
        status = success
        print(obj, success, lineno, u)
f.close()

print(status, lineno, "done")
```
